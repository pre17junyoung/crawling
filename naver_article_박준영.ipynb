```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()

url = "https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=105#&date=%2000:00:00&page=1"
```

```python
driver.get(url)
```

```python
titles = []
for i in range(10):
    
    datas = driver.find_elements_by_css_selector('#section_body li > dl > dt:nth-child(2) > a')

    for title in datas:
        titles.append(title.text)
        
    driver.find_elements_by_css_selector('#paging a')[i].click()
    time.sleep(1)
    
print(titles)
    
    
```
